```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_features=5000) # Limiting features to 5000 to manage dimensionality

# Ensure all values in 'v2' are strings before fitting the vectorizer
data['v2'] = data['v2'].astype(str)

# Fit and transform the 'v2' column (text messages)
X_tfidf = tfidf_vectorizer.fit_transform(data['v2'])

# Convert X_tfidf to a DataFrame for easier handling, if needed, though sklearn models can work with sparse matrices directly
X_tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Prepare the target variable
y = data['v1']

# Split the data into training and testing sets with the new TF-IDF features
X_train_tfidf, X_test_tfidf, y_train_tfidf, y_test_tfidf = train_test_split(X_tfidf_df, y, test_size=0.2, random_state=42)

print("Shape of X_train_tfidf:", X_train_tfidf.shape)
print("Shape of X_test_tfidf:", X_test_tfidf.shape)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create an SVM classifier with a linear kernel
svm_model_tfidf = SVC(kernel='linear', random_state=42)

# Train the model with TF-IDF features
svm_model_tfidf.fit(X_train_tfidf, y_train_tfidf)

# Make predictions on the training and test sets
y_train_pred_svm_tfidf = svm_model_tfidf.predict(X_train_tfidf)
y_test_pred_svm_tfidf = svm_model_tfidf.predict(X_test_tfidf)

# Evaluate the model
print("\n--- Support Vector Machine (SVM) Model Evaluation with TF-IDF Features ---")
print("Training Accuracy (SVM with TF-IDF):", accuracy_score(y_train_tfidf, y_train_pred_svm_tfidf))
print("Testing Accuracy (SVM with TF-IDF):", accuracy_score(y_test_tfidf, y_test_pred_svm_tfidf))
print("\nClassification Report (SVM with TF-IDF, Test Set):\n", classification_report(y_test_tfidf, y_test_pred_svm_tfidf))
print("\nConfusion Matrix (SVM with TF-IDF, Test Set):\n", confusion_matrix(y_test_tfidf, y_test_pred_svm_tfidf))
```
