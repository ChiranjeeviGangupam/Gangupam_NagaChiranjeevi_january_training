```python
num_columns=data.select_dtypes(include=['int64','float64']).columns.tolist()
num_columns
```

```python
# First, fill missing categorical values in the main `data` DataFrame with their mode.
for col in cat_columns:
  mode = data[col].mode()[0]
  data.loc[:, col] = data[col].fillna(mode)

# Create copies for demonstration of mean and median imputation.
# At this point, `data` has its categorical missing values filled.
mean_val = data.copy()
median_val = data.copy()

# Fill missing numerical values in the `mean_val` DataFrame with the mean of the original column.
for col in num_columns:
  # Using the mean of the column from the 'data' DataFrame (which at this point may still have NaNs in 'Rating')
  # or use `mean_val[col].mean()` if you want the mean of the data copied so far.
  # Sticking to `data[col].mean()` to get the mean from the state *before* this specific mean imputation.
  mean_val_col = data[col].mean()
  mean_val.loc[:, col] = mean_val[col].fillna(mean_val_col)

# Fill missing numerical values in the `median_val` DataFrame with the median of the original column.
for col in num_columns:
  median_val_col = data[col].median()
  median_val.loc[:, col] = median_val[col].fillna(median_val_col)

print("Data after filling categorical missing values with mode (original 'data' state after mode imputation):")
print(data.head())
print("\nData after filling numerical with Mean and categorical with Mode (mean_val):")
print(mean_val.head())
print("\nData after filling numerical with Median and categorical with Mode (median_val):")
print(median_val.head())
```

For a numerical column like 'Rating', both mean and median imputation are commonly used, while mode imputation is generally less suitable unless the data is discrete or has a very distinct peak.

Given that outliers were identified in the 'Rating' column earlier in the notebook, median imputation would generally be the more robust and therefore 'best' method to use. It minimizes the distortion of the data distribution due to extreme values if they were present before imputation. Since the mean and median of the 'Rating' column (after outlier removal, or considering the values before imputation) are usually quite close, both methods might yield similar results in some cases, but median offers better resilience against potential skewness or outliers.

 Detecting and treating outliers using IQR method






```python
outlier_info = {}

for col in num_columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)]
    outlier_info[col] = outliers.shape[0]

    print(f"{col}: {outliers.shape[0]} outliers")
```

```python
data = data[~((data['Rating'] < lower_bound) | (data['Rating'] > upper_bound))]
print(f"Shape of data after removing outliers: {data.shape}")
```


Which is most effective and why?

Without a specific machine learning model or objective, it's hard to definitively say which is most effective. However, here are some general considerations for your 'Rating' column (which is positive and had outliers removed):

Standard Scaling is often a good default choice, especially for algorithms that rely on distance metrics (like K-Nearest Neighbors, SVMs, or neural networks) because it ensures all features contribute equally to the distance calculation. It's also robust to some extent against remaining mild outliers.
Min-Max Scaling is also a strong contender, especially since you've already removed outliers. For a feature like 'Rating' which inherently has a bounded scale (e.g., 1-5), Min-Max scaling can be very intuitive and effective, explicitly putting all ratings on a consistent 0-1 scale.
Max Absolute Scaling might be less commonly chosen here unless the data naturally centered around zero or had a specific sparsity requirement, which is not typical for ratings.
Given that 'Rating' is a continuous, positive-valued numerical feature with outliers already addressed, both Min-Max Scaling and Standard Scaling are likely to perform well. Many practitioners would lean towards Standard Scaling as a generally robust first choice, but Min-Max Scaling can be equally, if not more, appropriate if the desired outcome is to have all ratings explicitly within a [0, 1] range.

In summary, for your dataset:

For features like Category and Genres (which are nominal but have varying numbers of unique values), Binary Encoding or Target Encoding are generally strong choices. Binary encoding balances dimensionality reduction with avoiding false ordinality. Target encoding can capture powerful predictive information if done correctly.
For features with few unique nominal categories (e.g., Type, Content Rating), One-Hot Encoding is often the safest and most interpretable choice.
For features with very high cardinality like App, Current Ver, Android Ver, or Last Updated, techniques like Binary Encoding, Frequency Encoding, or Target Encoding (if a clear target relationship exists) are preferable over One-Hot or Label Encoding to manage dimensionality and avoid artificial ordering. Label encoding was applied to these, but caution is advised if the numerical order doesn't reflect a true relationship.
The 'best' method ultimately depends on the specific machine learning model you plan to use and the inherent nature of the relationship between your categorical features and your target variable. Experimentation and model performance evaluation would be key to determining the most effective encoding for your specific task.
