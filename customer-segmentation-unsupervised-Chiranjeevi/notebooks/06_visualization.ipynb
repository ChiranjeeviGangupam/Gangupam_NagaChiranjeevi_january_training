```python

# Visualization 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import os

os.makedirs("../results/cluster_plots", exist_ok=True)
os.makedirs("../results/pca_outputs", exist_ok=True)

df = pd.read_csv("../results/final_clusters.csv")


df = df.reset_index(drop=True)

# -------------------------------
# Separate features
# -------------------------------
X = df.drop(columns=['Cluster'])

# -------------------------------
# Scaling
# -------------------------------
scaler = StandardScaler()
scaled_data = scaler.fit_transform(X)

# -------------------------------
# PCA

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])

pca_df['Cluster'] = df['Cluster'].values

# 1. PCA Plot

plt.figure()
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df)
plt.title("PCA Cluster Visualization")

plt.savefig("../results/pca_outputs/pca_clusters.png")
plt.close()


# 2. Cluster Distribution

plt.figure()
sns.countplot(x='Cluster', data=df)
plt.title("Cluster Distribution")

plt.savefig("../results/cluster_plots/cluster_distribution.png")
plt.close()


# 3. Boxplots

features_to_plot = ['Recency', 'Frequency', 'Monetary']

for feature in features_to_plot:
    if feature in df.columns: 
        plt.figure()
        sns.boxplot(x='Cluster', y=feature, data=df)
        plt.title(f"{feature} by Cluster")

        plt.savefig(f"../results/cluster_plots/{feature}_boxplot.png")
        plt.close()


# 4. Heatmap

cluster_summary = df.groupby('Cluster').mean()

plt.figure(figsize=(8,6))
sns.heatmap(cluster_summary, annot=True, cmap='coolwarm')
plt.title("Cluster Heatmap")

plt.savefig("../results/cluster_plots/cluster_heatmap.png")
plt.close()

print(" Visualization completed and saved successfully!")
```
