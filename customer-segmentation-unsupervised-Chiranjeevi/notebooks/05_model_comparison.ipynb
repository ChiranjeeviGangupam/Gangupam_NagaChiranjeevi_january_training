```python

# Model Comparison & Best Cluster Selection

import pandas as pd
import numpy as np

from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# Load data

df = pd.read_csv(r"C:\Users\gangu\Desktop\Project-demo\customer-segmentation-unsupervised\data\processed\processed_customer_data.csv", index_col=0)

# Scale data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Function to evaluate model

def evaluate_model(name, labels, data):
    unique_labels = set(labels)

    # Avoid invalid cases (DBSCAN noise etc.)
    if len(unique_labels) <= 1 or -1 in unique_labels:
        return {
            "Model": name,
            "Clusters": len(unique_labels),
            "Silhouette": -1,
            "Davies_Bouldin": -1
        }

    return {
        "Model": name,
        "Clusters": len(unique_labels),
        "Silhouette": round(silhouette_score(data, labels), 4),
        "Davies_Bouldin": round(davies_bouldin_score(data, labels), 4)
    }


# Run all models

results = []

# KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
k_labels = kmeans.fit_predict(scaled_data)
results.append(evaluate_model("KMeans", k_labels, scaled_data))

# Hierarchical
hier = AgglomerativeClustering(n_clusters=4)
h_labels = hier.fit_predict(scaled_data)
results.append(evaluate_model("Hierarchical", h_labels, scaled_data))

# DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
d_labels = dbscan.fit_predict(scaled_data)
results.append(evaluate_model("DBSCAN", d_labels, scaled_data))

# GMM
gmm = GaussianMixture(n_components=4, random_state=42)
g_labels = gmm.fit_predict(scaled_data)
results.append(evaluate_model("GMM", g_labels, scaled_data))

# Create comparison table


results_df = pd.DataFrame(results)

print("\nModel Comparison Results:\n")
print(results_df)


# Select best model


# Criteria:
# - High Silhouette Score
# - Low Davies-Bouldin Index

best_model = results_df.sort_values(
    by=["Silhouette", "Davies_Bouldin"],
    ascending=[False, True]
).iloc[0]

print("\nBest Model Selected:")
print(best_model)
```
