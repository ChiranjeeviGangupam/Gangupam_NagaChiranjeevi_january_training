```python

# Exploratory Data Analysis (EDA)


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load processed dataset
df = pd.read_csv("C:\\Users\\gangu\\Desktop\\Project-demo\\customer-segmentation-unsupervised\\data\\processed\\processed_customer_data.csv")

print("Dataset shape:", df.shape)
print(df.describe())

# Set style (for better visuals)
sns.set(style="whitegrid")

# 1. UNIVARIATE ANALYSIS


# Distribution of Recency
plt.figure()
sns.histplot(df['Recency'], kde=True)
plt.title("Recency Distribution")
plt.xlabel("Days since last purchase")
plt.show()

# Distribution of Frequency
plt.figure()
sns.histplot(df['Frequency'], kde=True)
plt.title("Frequency Distribution")
plt.show()

# Distribution of Monetary
plt.figure()
sns.histplot(df['Monetary'], kde=True)
plt.title("Monetary Distribution")
plt.show()


# 2. BIVARIATE ANALYSIS


# Recency vs Monetary
plt.figure()
sns.scatterplot(x='Recency', y='Monetary', data=df)
plt.title("Recency vs Monetary")
plt.show()

# Frequency vs Monetary
plt.figure()
sns.scatterplot(x='Frequency', y='Monetary', data=df)
plt.title("Frequency vs Monetary")
plt.show()

# Avg Order Value vs Frequency
plt.figure()
sns.scatterplot(x='Frequency', y='Avg_Order_Value', data=df)
plt.title("Frequency vs Avg Order Value")
plt.show()

# 3. MULTIVARIATE ANALYSIS


# Pairplot (important for clustering understanding)
sns.pairplot(df[['Recency', 'Frequency', 'Monetary']])
plt.show()


```

```python
# 4. CORRELATION HEATMAP


plt.figure(figsize=(8,6))
corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python

# 5. PATTERN DISCOVERY


# High value customers
high_value = df[df['Monetary'] > df['Monetary'].quantile(0.75)]
print("High value customers:", high_value.shape[0])

# Low frequency customers
low_freq = df[df['Frequency'] < df['Frequency'].quantile(0.25)]
print("Low frequency customers:", low_freq.shape[0])

# Recent customers
recent = df[df['Recency'] < df['Recency'].quantile(0.25)]
print("Recent customers:", recent.shape[0])
```
